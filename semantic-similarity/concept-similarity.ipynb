```python
import nltk
#nltk.download('wordnet')
from nltk.corpus import wordnet as wn
import os
import csv
```

Lettura del file 'WordSim353.csv'

```python
current_dir = os.getcwd()
file_name = 'WordSim353.csv'

with open(file_name, 'r', encoding='utf-8') as train:
   righe = train.readlines()[1:]
   print(righe)
```

```python
def get_synset_depth(synset):
    hypernym_paths = synset.hypernym_paths()
    if not hypernym_paths:
        return 0
    return max(len(path) for path in hypernym_paths) - 1

# per capire: stampa tutti i synsets (nodi) in cui si trova un certo termine, poi per ogni synset stampa la sua definizione e i termini che ne fanno parte
for riga in righe:
    riga = riga.split(',')
    term1 = riga[0]
    term2 = riga[1]
    print(term1, term2)
    synsets1 = wn.synsets(term1)
    print('\nSynsets for lemma "{}: "'.format(term1))
    print(synsets1)
    for synset in synsets1:
        print('\nSynset name: {}\n'.format(synset.name()))
        print('Synset\'s depth: {}\n'.format(get_synset_depth(synset)))
        print('Lemmas for the synset: {}\n'.format(synset.lemmas()))
        print('Definition: {}\n--------'.format(synset.definition()))  
```

### I valori di confronto

```python
#valori ottenuti con le funzioni di 
print('Expected values:\n')
for riga in righe:
    riga = riga.split(',')
    term1 = riga[0]
    term2 = riga[1]
    s1 = wn.synsets(term1)[0]
    s2 = wn.synsets(term2)[0]
    print(term1, term2)
    print(s1,s2)
    print('wup_similarity({}, {}) = {}'.format(s1, s2, s1.wup_similarity(s2)))
    print('lch_similarity({}, {}) = {}'.format(s1, s2, s1.lch_similarity(s2)))
    print('path_similarity({}, {}) = {}\n\n'.format(s1, s2, s1.path_similarity(s2)))
```
